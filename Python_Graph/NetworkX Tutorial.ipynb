# Create a graph with NetworkX 
- https://pypi.org/project/networkx/

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_edge('A', 'B', weight=4)
G.add_edge('B', 'D', weight=2)
G.add_edge('A', 'C', weight=3)
G.add_edge('C', 'D', weight=4)
nx.shortest_path(G, 'A', 'D', weight='weight')
```

## Draw Graph

```python
nx.draw(G)
plt.show()
```

```python
G = nx.Graph()
G.add_edges_from([(1 ,2) , (2 ,3) , (1 ,3) , (1 ,4) ]) # add edges from a list of vertices
pos = { 1: (20, 30), 2: (40, 30), 3: (30, 10),4: (0, 40)} # draw by position

nx.draw_networkx(G, pos=pos)

plt.show()
```

```python
FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for n, nbrs in FG.adj.items():
    for nbr, eattr in nbrs.items():
        wt = eattr['weight']
        if wt < 0.5: print(f"({n}, {nbr}, {wt:.3})")
       
```

```python
nx.draw_networkx(FG)

plt.show()
```

```python
pos=nx.spring_layout(FG) # pos = nx.nx_agraph.graphviz_layout(G)
nx.draw_networkx(FG,pos)
labels = nx.get_edge_attributes(FG,'weight')
nx.draw_networkx_edge_labels(FG,pos,edge_labels=labels)
```
