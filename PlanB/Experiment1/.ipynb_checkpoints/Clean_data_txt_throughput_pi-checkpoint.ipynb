```python
## function to get dataframe from df 

def getDataframeThru(df,start_row,measurement_interval,header_range):
    '''
    measurement_interval = 30 (sec) : 
    header_range = 9 lines
    start_row = 0 
    '''
    
    df1 = df.drop(labels=range(start_row, header_range), axis=0)
    df1t = df1.drop(labels=range(measurement_interval, len(df)), axis=0)
    
    return df1t
```

```python
skip_1 = list(range(0, 10, 1))
df = pd.read_csv(filename,
   skiprows=skip_1,
   header=None,
   delimiter=' ',
   skipinitialspace=True,
   error_bad_lines=False)
```

```python
print(df)
```

```python
h = 9
measurement_interval = 30 
 
start_row = 0 
df_all = df.drop(labels=range(1, len(df)), axis=0)
print(df_all)
for i in range(0,2):
    print('i = ',i)
    if i == 0 :
        m = 30 + h
        df1 = getDataframeThru(df,start_row,m,0)
        result = pd.concat([df_all,df1])
        df_all = df.drop(labels=0, axis=0)
    else:
        header_range = measurement_interval + header_range -1
        measurement_interval = measurement_interval + header_range -1
        print('measurement_interval: ',measurement_interval)
        df1 = getDataframeThru(df,start_row,measurement_interval,header_range)
        print('df1 = ',df1)
        result = pd.concat([df_all,df1])
        print('hd = ', header_range)
    df_all = result
```
